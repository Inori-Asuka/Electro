```python
import cv2
import matplotlib.pyplot as plt
import os

def apply_and_show_mask(img_path, mask_path):

    img = cv2.imread(img_path)
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)

    if mask.shape != img.shape[:2]:
        mask = cv2.resize(mask, (img.shape[1], img.shape[0]))

    masked_img = cv2.bitwise_and(img, img, mask=mask)

    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    masked_img_rgb = cv2.cvtColor(masked_img, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(15, 5))

    # 显示原图
    plt.subplot(1, 3, 1)
    plt.title("Original Image ")
    plt.imshow(img_rgb)
    plt.axis('off')

    # 显示 Mask
    plt.subplot(1, 3, 2)
    plt.title("Mask")
    plt.imshow(mask, cmap='gray')
    plt.axis('off')

    # 显示结果
    plt.subplot(1, 3, 3)
    plt.title("Masked Result")
    plt.imshow(masked_img_rgb)
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
file_path_img = '/data/wly/data/eletric/1/1_c.png'
file_path_mask = '/data/wly/data/eletric/1/1_c_mask.png'

apply_and_show_mask(file_path_img, file_path_mask)
```
